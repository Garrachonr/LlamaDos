```python
# Let's look at original dataset
main = pd.read_csv('DAILYD_translation_en2es.csv')
main.head(10)

main.head(10)
```

```python

count = 000000
id = 0
buffer = []
system_prompt = "You are a helpful, respectful and honest conversational assistant. Have a conversation with the user in a natural way.  Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature."
all_conversations = []

for i in main["UID"]:

  #Identificador para diferenciar distitnas conversaciones
  if i[7:13] != count:
    buffer = []
    count = i[7:13]
    id = id + 1
  # if float(count) >= 13116:
  #   break
  buffer.append(list(main.loc[(main.UID == i)].translation)[0])
  if len(buffer) == 2:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s>".format(system_prompt, buffer[0], buffer[1])
    all_conversations.append({"text": text})
  elif len(buffer) == 3:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST]".format(system_prompt, buffer[0], buffer[1], buffer[2])
    all_conversations.append({"text": text})
  elif len(buffer) == 4:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s>".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3])
    all_conversations.append({"text": text})
  elif len(buffer) == 5:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST]".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4])
    all_conversations.append({"text": text})
  elif len(buffer) == 6:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s>".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4], buffer[5])
    all_conversations.append({"text": text})
  elif len(buffer) == 7:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST]".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4], buffer[5], buffer[6])
    all_conversations.append({"text": text})
  elif len(buffer) == 8:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s>".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4], buffer[5], buffer[6], buffer[7])
    all_conversations.append({"text": text})
  elif len(buffer) == 9:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST]".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4], buffer[5], buffer[6], buffer[7], buffer[8])
    all_conversations.append({"text": text})
  elif len(buffer) == 10:
    text = "<s>[INST] <<SYS>> {} <</SYS>> {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s><s>[INST] {} [/INST] {} </s>".format(system_prompt, buffer[0], buffer[1], buffer[2], buffer[3], buffer[4], buffer[5], buffer[6], buffer[7], buffer[8], buffer[9])
    all_conversations.append({"text": text})

file_path = 'DailyDB-Spanish-Llama2.json'

with open(file_path, 'w') as file:
    for item in all_conversations:
        json.dump(item, file)  # Write the JSON object to the file
        file.write('\n')       # Add a newline after each object

print("Archivo JSON con todas las conversaciones generado exitosamente.")
```
